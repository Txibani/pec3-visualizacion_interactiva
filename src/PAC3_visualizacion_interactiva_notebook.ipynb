```python
### Limpieza y agrupamiento de los datos
```

```python
# Se renombran las variables código de comarca, fecha y sexos para los datasets
# de casos de covid y defunciones por covid

# Dataset registros
df_cases = df_cases.drop(['ComarcaCodi', 'MunicipiCodi', 'MunicipiDescripcio', 'DistricteCodi',
                          'DistricteDescripcio', 'SexeCodi', 'TipusCasDescripcio'], axis=1)
df_cases.columns = ['data', 'comarca', 'sexe_casos', 'num_casos']
df_cases['sexe_casos'] = df_cases['sexe_casos'].str.lower()

# Dataset fallecimientos
df_deaths = df_deaths.drop(['Codi Comarca', 'Codi Sexe'], axis=1)
df_deaths.columns = ['data', 'comarca', 'sexe_defuncions', 'num_defuncions']
df_cases['sexe_defuncions'] = df_deaths['sexe_defuncions'].str.lower()
```

```python
# DATASET CASOS DE COVID

# Agrupamos numéros de casos por mes, comarca y sexo
df_cases['data'] = pd.to_datetime(df_cases['data'], format='%d/%m/%Y')

df_months_cases = df_cases.groupby([(df_cases.data.dt.month.values), (df_cases.data.dt.year.values),
                                    'comarca', 'sexe_casos'])['num_casos'].sum().reset_index()
df_months_cases
# Renombramos las variables creadas
df_months_cases.rename(columns={'level_0': 'year'}, inplace = True)
df_months_cases.rename(columns={'level_1': 'month'}, inplace = True)
# Concatenamos ambas variables en una nueva que contendrá el valor mes/año
df_months_cases['data'] = df_months_cases['month'].astype(str) + '/' + df_months_cases['year'].astype(str)
# Se borran las variables year y month que han sido concatenadas
df_months_cases.drop(['year', 'month'], inplace = True, axis = 1)

# Se transforma la tabla para generar un dataset cuyas columnas sean la fecha por mes y los
# casos por genéro
df_cases_sex = df_months_cases.copy()
df_cases_sex = df_cases_sex.pivot_table(index = ['data'],
                                        columns = ['sexe_casos'], aggfunc='sum').reset_index('data')
df_cases_sex.columns = [c[0] + "_" + c[1] for c in df_cases_sex.columns]
df_cases_sex['data_'] = pd.to_datetime(df_cases_sex['data_'], format='%Y/%m')
df_cases_sex.sort_values(by = 'data_', inplace = True)
df_cases_sex['data_'] = df_cases_sex['data_'].dt.strftime('%m/%Y')
df_cases_sex
```

```python
# DATASET FALLECIMIENTOS POR COVID

# Agrupamos numéros de fallecimientsos por mes, comarca y sexo
df_deaths['data'] = pd.to_datetime(df_deaths['data'], format='%d/%m/%Y')
df_months_deaths = df_deaths.groupby([(df_deaths.data.dt.month.values), (df_deaths.data.dt.year.values),
                                    'comarca', 'sexe_defuncions'])['num_defuncions'].sum().reset_index()
# Renombramos las variables creadas
df_months_deaths.rename(columns={'level_0': 'year'}, inplace = True)
df_months_deaths.rename(columns={'level_1': 'month'}, inplace = True)
# Concatenamos ambas variables en una nueva que contendrá el valor mes/año
df_months_deaths['data'] = df_months_deaths['month'].astype(str) + '/' + df_months_deaths['year'].astype(str)
# Se borran las variables year y month que han sido concatenadas
df_months_deaths.drop(['year', 'month'], inplace = True, axis = 1)

# Se transforma la tabla para generar un dataset cuyas columnas sean la fecha por mes y los
# fallecimientos por genéro
df_deaths_sex = df_months_deaths.copy()
df_deaths_sex = df_deaths_sex.pivot_table(index='data', columns = ['sexe_defuncions'],
                                          aggfunc='sum').reset_index('data')
df_deaths_sex.columns = [c[0] + "_" + c[1] for c in df_deaths_sex.columns]
df_deaths_sex['data_'] = pd.to_datetime(df_deaths_sex['data_'], format='%Y/%m')
df_deaths_sex.sort_values(by = 'data_', inplace = True)
df_deaths_sex['data_'] = df_deaths_sex['data_'].dt.strftime('%m/%Y')
df_deaths_sex
```

```python
# DATASET VACUNAS

# Se convierte la columna DATA a tipo fecha
df_vaccines['DATA'] = pd.to_datetime(df_vaccines['DATA'], format='%d/%m/%Y')

# Se agrupa el dataset por fecha y fabricante para conseguir el total de vacunas por tipo
df_months_vaccines = df_vaccines.groupby([(df_vaccines.DATA.dt.month.values), (df_vaccines.DATA.dt.year.values),
                                    'FABRICANT'])['RECOMPTE'].sum().reset_index()

# Renombramos las variables creadas
df_months_vaccines.rename(columns={'level_0': 'month'}, inplace = True)
df_months_vaccines.rename(columns={'level_1': 'year'}, inplace = True)
# Concatenamos ambas variables en una nueva que contendrá el valor mes/año
df_months_vaccines['data'] = df_months_vaccines['month'].astype(str) + '/' + df_months_vaccines['year'].astype(str)
# Se borran las variables year y month que han sido concatenadas
df_months_vaccines.drop(['year', 'month'], inplace = True, axis = 1)
df_months_vaccines

# Se transforma la tabla para generar un dataset cuyas columnas sean la fecha y los
# distintos tipos de vacuna administradas por mes, ordeandas en el tiempo
df_vaccines_fabricant = df_months_vaccines.copy()
df_vaccines_fabricant = df_vaccines_fabricant.pivot_table(index = ['data'],
                                        columns = ['FABRICANT'], aggfunc='sum').reset_index('data')
# Se renombran las variables y se ordena el dataset por fecha                                        
df_vaccines_fabricant.columns = [c[0] + "_" + c[1] for c in df_vaccines_fabricant.columns]
df_vaccines_fabricant['data_'] = pd.to_datetime(df_vaccines_fabricant['data_'], format='%m/%Y')
df_vaccines_fabricant.sort_values(by = 'data_', inplace = True)
df_vaccines_fabricant['data_'] = df_vaccines_fabricant['data_'].dt.strftime('%m/%Y')
df_vaccines_fabricant
```

##### Dataset resultante de la unión de dataset Registre_de_casos_de_COVID-19_a_Catalunya_per_municipi_i_sexe y registre_de_defuncions_per_COVID-19_a_Catalunya_per_comarca_i_sexe

```python
# Obtenemos el números de casos total en base a las variables data, comarca y sexe
df_cases = df_cases.groupby(['data', 'comarca', 'sexe_casos'])['num_casos'].sum().reset_index()

df_deaths = df_deaths.groupby(['data', 'comarca', 'sexe_defuncions'])['num_defuncions'].sum().reset_index()
```

```python
# Binarizamos las variables sexe para el dataframe df_cases mediante la aplicación de one-hot enconding
from sklearn.preprocessing import OneHotEncoder
onehot_encoder = OneHotEncoder()

# Aplicamos one-hot encoder a la variable estado sexe_casos
sexe_encoded_df = pd.DataFrame(onehot_encoder.fit_transform(df_cases[['sexe_casos']]).toarray())
sexe_encoded_df.columns = ['dona', 'home']

# Se borran la variable sexe que no ha sido binarizadas
df_cases.drop('sexe_casos', inplace = True, axis = 1)

# Juntamos todos los datasets para general uno final
df_cases = pd.concat([df_cases, sexe_encoded_df], axis = 1)

# Renombramos las columnas credas para diferenciarlas en el dataframe final
df_cases.rename(columns={'dona': 'dona_cases'}, inplace = True)
df_cases.rename(columns={'home': 'home_cases'}, inplace = True)
```

```python
# Binarizamos las variables sexe para el dataframe df_deaths mediante la aplicación de one-hot enconding
from sklearn.preprocessing import OneHotEncoder
onehot_encoder = OneHotEncoder()

# Aplicamos one-hot encoder a la variable estado sexe_defuncions
sexe_encoded_df = pd.DataFrame(onehot_encoder.fit_transform(df_deaths[['sexe_defuncions']]).toarray())
sexe_encoded_df.columns = ['dona', 'home', 'No classificat']

# Se borran la variable sexe que no ha sido binarizadas
df_deaths.drop('sexe_defuncions', inplace = True, axis = 1)

# Juntamos todos los datasets para general uno final
df_deaths = pd.concat([df_deaths, sexe_encoded_df], axis = 1)

# Renombramos las columnas credas para diferenciarlas en el dataframe final
df_deaths.rename(columns={'dona': 'dona_deaths'}, inplace = True)
df_deaths.rename(columns={'home': 'home_deaths'}, inplace = True)
df_cases.rename(columns={'No classificat': 'no_classificat_deaths'}, inplace = True)
```

```python
# Juntamos ambos datasets en uno solo
final_df = df_cases.merge(df_deaths, how = 'left', left_on = ['data', 'comarca'], right_on = ['data', 'comarca'])
# Se ordena el dataset for fecha
final_df['data'] = pd.to_datetime(final_df['data'], format='%d/%m/%Y')
final_df.sort_values(by = 'data', inplace = True)
final_df['data'] = final_df['data'].dt.strftime('%d/%m/%Y')
final_df.head(5)
```

```python
# Renombramos las variables de los distintos datasets para que sean más intuitivas
final_df.columns = ['fecha', 'Comarca', 'Número de casos total', 'Casos positivos de mujeres', 'Casos positivos de hombres',
                    'Número de defunciones total', 'Defunciones de mujeres', 'Defunciones de hombres',
                    'Defunciones no clasificadas']

df_cases_sex.columns = ['Mes', 'Número de positivos en mujeres', 'Número de positivos en hombres']

df_deaths_sex.columns = ['Mes', 'Número de defunciones de mujeres', 'Número de defunciones de hombres',
                         'Número de defunciones no clasificadas']
```
